```python
value=("""My name is prince.
My best friends are : 
Dhruv,
Dev, 
Rajat, 
Partik, 
Sachin, 
Chetan, 
Narendar, 
Krishna, 
Sumit,
Harsh and 
Mukesh.
Total 12 friends I have. """)
with open('C:/Users/0526p/Documents/experiment.txt','w') as f:
    s=str(value)
    f.write(s)
```

```python
enter='My friend rajat birthday is ', 6, 'February ', 1999
with open('C:/Users/0526p/Documents/experiment.txt','a') as f:
    s=str(enter)
    f.write('\n')
    f.write('\n')
    f.write(s)
with open('C:/Users/0526p/Documents/experiment.txt','r') as f:
    for line in f:
        print(line, end='')
```

```python
# a+ is for append text and read documents.

with open('C:/Users/0526p/Documents/experiment.txt','a+') as f:
    f.seek(4)
    f.write('\n')
    f.write('\n')
    f.write('my birthday date is 26 may 1999')
```
